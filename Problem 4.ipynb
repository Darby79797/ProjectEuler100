A solution to Problem 4 of Project Euler.

Find the largest palindrome made from the product of two 3-digit numbers.

We just list them, and will optimise further only if the code doesn't run quickly.

```python
def isPalindrome(n):
    if str(n) == str(n)[::-1]:
        return True
    else:
        return False
```

```python
for i in range(100,999):
    for j in range(100,999):
        if i>=j:
            if isPalindrome(i*j):
                print(i*j)
print('Done!')
```

Ahh, 906609 looks like it'll probably be the biggest. It is.
